```python


#Uploading data from drive to notebook
#Data was collected from Ercot Data Portal

solar_data_2024_file_id = "1yq8PIsaBdRqjDZwBtU8DXtUkyOZ_A-MW"
url = f'https://drive.google.com/uc?id={solar_data_2024_file_id}'

output = 'solar_data_2024.xlsx'
gdown.download(url, output, quiet=False)

solar_data_2024_df = pd.read_excel(output)

solar_data_2023_file_id = "1w8yGZ5lIZAsV2QastEmoezy64lYsQTrS"
url = f'https://drive.google.com/uc?id={solar_data_2023_file_id}'

output = 'solar_data_2023.xlsx'
gdown.download(url, output, quiet=False)

solar_data_2023_df = pd.read_excel(output)

solar_data_2022_file_id = "16wbRUPOzEOzUEA9_CeO6JnqMkL4EM5L0"
url = f'https://drive.google.com/uc?id={solar_data_2022_file_id}'

output = 'solar_data_2022.xlsx'
gdown.download(url, output, quiet=False)

solar_data_2022_df = pd.read_excel(output)

wind_data_2022_file_id = "14VKyrH9-XQrsYd3RvY5YfZOb9lwehZt7"
url = f'https://drive.google.com/uc?id={wind_data_2022_file_id}'

output = 'wind_data_2022.xlsx'
gdown.download(url, output, quiet=False)

wind_data_2022_df = pd.read_excel(output)

wind_data_2023_file_id = "1urT9E3QFX-xQQ7mTrioPBrXx3KmiTK35"
url = f'https://drive.google.com/uc?id={wind_data_2023_file_id}'

output = 'wind_data_2023.xlsx'
gdown.download(url, output, quiet=False)

wind_data_2023_df = pd.read_excel(output)

wind_data_2024_file_id = "1_XfCSTREiiU5RbjiRznw4hhCC5AZt8R4"
url = f'https://drive.google.com/uc?id={wind_data_2024_file_id}'

output = 'wind_data_2024.xlsx'
gdown.download(url, output, quiet=False)

wind_data_2024_df = pd.read_excel(output)
```

```python
solar_data_2022_to_2024_df = pd.concat(
    [solar_data_2022_df, solar_data_2023_df, solar_data_2024_df],
    ignore_index=True
)
```

```python
wind_data_2022_to_2024_df = pd.concat(
    [wind_data_2022_df, wind_data_2023_df, wind_data_2024_df],
    ignore_index=True
)
```

```python
# This code merges wind and solar datasets from 2022 to 2024 based on the 'Time (Hour-Ending)' column.
# It adds suffixes to distinguish overlapping column names (e.g., '_wind' and '_solar').
# The last line displays all but the final 5 rows of the merged DataFrame.

merged_solar_and_wind_df = pd.merge(
    wind_data_2022_to_2024_df,
    solar_data_2022_to_2024_df,
    on=['Time (Hour-Ending)'],
    suffixes=('_wind', '_solar')
)

merged_solar_and_wind_df.head(-5)
```

```python
columns_to_drop = [
    'Date_wind',
    'Date_solar',
    'Wind Output, % of Load',
    'Solar Output, % of Load',
    'Wind Output, % of Installed',
    'Solar Output, % of Installed',
    'Daytime Hour',
    'Ramping Daytime Hour'
]

merged_solar_and_wind_df.drop(columns=columns_to_drop, inplace=True)

merged_solar_and_wind_df
```

```python
# Check for duplicate rows in the DataFrame
duplicates = merged_solar_and_wind_df[merged_solar_and_wind_df.duplicated(keep=False)]

# If there are any duplicates, display them
if not duplicates.empty:
    print(f"Duplicate rows found:\n{duplicates}")
else:
    print("No duplicates found.")
```

```python
columns_to_drop = [
    'Total Wind Installed, MW',
    'Total Solar Installed, MW'
]

merged_solar_and_wind_df.drop(columns=columns_to_drop, inplace=True)

merged_solar_and_wind_df
```

```python
merged_solar_and_wind_df.to_csv('wind_solar_data.csv', index=False)
```
